## Example workflow

```python
# Step 0: import matplotlib and get it ready 
%matplotlib inline
import matplotlib.pyplot as plt

# Step 1: Prepare your data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Step 2: Set up your plot
fig, ax = plt.subplots(figsize = (10, 10)) # (width, height)

# Step 3: Plot the data
ax.plot(x, y)

# Step 4: Customize your plot
ax.set(title = "Your first plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# Step 5: Save and show
fig.savefig('sample-plot.png')
```

## Options for subplots

```python
# Option 1:

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5))

# Plotting to each different axis
ax1.plot(x, x / 2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2: 

fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))

# Plotting to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### PyPlot vs Matplotlib OO method: Which to choose?

* Something quick, use pyplot method
* Something advanced, use OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind = 'scatter', 
             x = 'age', 
             y = 'chol',
            c = 'target',
            figsize = (10, 5));
```

```python
# OO method (kind of) mixed with PyPlot method
fig, ax = plt.subplots(figsize = (10, 5))

over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax); 

# ax.set_xlim([45, 100])
```

```python
over_50.target.values
```

```python
## OO method from scratch

fig, ax = plt.subplots(figsize = (10, 5))

# Plotting the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"]);

# Customizing the plot
ax.set(title = "Heart disease and cholesterol levels",
      xlabel = "Age",
      ylabel = "Cholesterol");

# Adding a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Adding a horizontal line (for the average)
ax.axhline(over_50["chol"].mean(),
          linestyle = '--');
```

```python
over_50.head()
```

```python
# Subplotting cholesterol, age, heart rate
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex = True)

# Adding data to ax0
scatter = ax0.scatter(x = over_50['age'], 
                      y = over_50['chol'], 
                      c = over_50['target'])

# Customizing ax0 
ax0.set(title = "Heart disease and cholesterol levels",
       ylabel = "Cholesterol")

# Adding a legend to ax0 
ax0.legend(*scatter.legend_elements(), title = "Target");

# Adding a meanline
ax0.axhline(y = over_50["chol"].mean(), linestyle = "--");

# Adding data to ax1
scatter = ax1.scatter(x = over_50['age'], 
                      y = over_50['thalach'], 
                      c = over_50['target']); 

# Customizing ax1
ax1.set(title = "Heart Disease and Max Hearth Rate",
       xlabel = "Age",
       ylabel = "Max heart rate");

# Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target");

# Adding a meanline
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# Adding a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

## OO method from scratch

fig, ax = plt.subplots(figsize = (10, 5))

# Plotting the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"],
                    cmap = 'summer'); # this changes the colour scheme

# Customizing the plot
ax.set(title = "Heart disease and cholesterol levels",
      xlabel = "Age",
      ylabel = "Cholesterol");

# Adding a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Adding a horizontal line (for the average)
ax.axhline(over_50["chol"].mean(),
          linestyle = '--');
```

```python
# Customizing y and x axis limitations:

# Subplotting cholesterol, age, heart rate
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                              sharex = True)

# Adding data to ax0
scatter = ax0.scatter(x = over_50['age'], 
                      y = over_50['chol'], 
                      c = over_50['target'],
                     cmap = 'winter')

# Customizing ax0 
ax0.set(title = "Heart disease and cholesterol levels",
       ylabel = "Cholesterol")

# Changing the x axis limits
ax0.set_xlim([50, 80])

# Adding a legend to ax0 
ax0.legend(*scatter.legend_elements(), title = "Target");

# Adding a meanline
ax0.axhline(y = over_50["chol"].mean(), linestyle = "--");

# Adding data to ax1
scatter = ax1.scatter(x = over_50['age'], 
                      y = over_50['thalach'], 
                      c = over_50['target'],
                     cmap = 'winter'); 

# Customizing ax1
ax1.set(title = "Heart Disease and Max Hearth Rate",
       xlabel = "Age",
       ylabel = "Max heart rate");

# Changing ax1 x axis limits
ax1.set_xlim([50, 80])

ax1.set_ylim([60, 200])

# Adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target");

# Adding a meanline
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# Adding a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
